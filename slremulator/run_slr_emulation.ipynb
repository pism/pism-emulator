```python
    fig = plt.figure()
    ax = fig.add_subplot(111)

    sns.distplot(
        q_as19,
        bins=bins,
        hist=False,
        hist_kws={"alpha": 0.3},
        norm_hist=True,
        kde=True,
        kde_kws={"shade": False, "alpha": 1.0, "linewidth": 1.5},
        ax=ax,
        color=cmap[-1],
        label=f"AS19 {kl_divergence(P, Q_as19):2.3f}",
    )
    m_qc = []
    m_yqc = []
    for k, method in enumerate(["expquad"]):

        kld = les_stats_df_500_gp.kldiv.values[::2][k]
        q = les_df_500[les_df_500.method == method]["Y_mean"].values
        q_kernel = stats.gaussian_kde(np.squeeze(q), bw_method=2 / q.std(ddof=1))

        pctl = 95
        m_qc.append(np.percentile(q, pctl))
        m_yqc.append(q_kernel.evaluate(m_qc[k]))

        sns.distplot(
            q,
            bins=bins,
            hist=False,
            hist_kws={"alpha": 0.3},
            norm_hist=True,
            kde=True,
            kde_kws={"shade": False, "alpha": 1.0, "linewidth": 1.5},
            ax=ax,
            color=cmap2[k],
            label=f"GP-{method}, {kld:2.3f}",
        )

    sns.distplot(
        p,
        bins=bins,
        hist=False,
        hist_kws={"alpha": 0.3},
        norm_hist=True,
        kde=True,
        kde_kws={"shade": False, "alpha": 1.0, "linewidth": 1.5},
        ax=ax,
        color="k",
        label='"True"',
    )

    ax.set_xlim(-12, 62)
    ax.set_ylabel("Probability (1)")
    ax.set_xlabel("Sea-level contribution (cm sea-level equivalent)")
    sns.despine()
    legend = ax.legend(ncol=1)
    legend.get_frame().set_linewidth(0.0)
    legend.get_frame().set_alpha(0.0)
    legend.remove()
    set_size(3.2, 3.2 / golden_ratio)

    for k, method in enumerate(["expquad"]):

        qc = m_qc[k]
        yqc = m_yqc[k]
        ax.plot(
            [qc, qc], [0, yqc], linewidth=1.25, linestyle="dotted", color=cmap2[k], alpha=1.0, transform=ax.transData
        )

    pc = np.percentile(p, pctl)
    p_kernel = stats.gaussian_kde(np.squeeze(p), bw_method=2 / p.std(ddof=1))
    ypc = p_kernel.evaluate(pc)
    ax.plot([pc, pc], [0, ypc], linewidth=1.25, linestyle="dotted", color="k", alpha=1.0, transform=ax.transData)

    qc_as19 = np.percentile(q_as19, pctl)
    q_as19_kernel = stats.gaussian_kde(np.squeeze(q_as19), bw_method=2 / q_as19.std(ddof=1))
    ypc_as19 = p_kernel.evaluate(qc_as19)
    ax.plot(
        [qc_as19, qc_as19],
        [0, ypc_as19],
        linewidth=1.25,
        linestyle="dotted",
        color=cmap[-1],
        alpha=1.0,
        transform=ax.transData,
    )

    ax.set_xlim(-12, 62)
    ax.set_ylabel("Probability (1)")
    ax.set_xlabel("Sea-level contribution (cm sea-level equivalent)")
    sns.despine()
    legend = ax.legend(ncol=1)
    legend.get_frame().set_linewidth(0.0)
    legend.get_frame().set_alpha(0.0)
    legend.remove()
    set_size(3.2, 3.2 / golden_ratio)

    outfile = "gp_emulator_best.pdf"
    print("Saving figure to {}".format(outfile))
    fig.savefig(outfile, bbox_inches="tight")
```
